## Critics Splats general info

```python
def critic_splat_general_info(url):
    response_URL_critic_splat = requests.get(url)
    soup_critic_splat = BeautifulSoup(response_URL_critic_splat.text)
    critic_splat_title = soup_critic_splat.find_all(slot='title')[13] #title
    critic_splat_title_clean = [title.text.strip() for title in critic_splat_title]
    critic_splat_info = soup_critic_splat.find_all(slot='metadataProp') #general info
    critic_splat_info_clean = [info.text.strip() for info in critic_splat_info]
    critic_splat_genre = soup_critic_splat.find_all(slot='metadataGenre') #genre
    critic_splat_genre_clean = [genre.text.strip() for genre in critic_splat_genre]
    critic_splat_critic_score = soup_critic_splat.find_all(slot='criticsScore')[0] #critic score
    critic_splat_critic_score_clean = [score.text.strip() for score in critic_splat_critic_score]
    critic_splat_critic_reviews = soup_critic_splat.find_all(slot='criticsReviews') # no. critic reviews
    critic_splat_critic_reviews_clean = [score.text.strip() for score in critic_splat_critic_reviews]
    critic_splat_audience_score = soup_critic_splat.find_all(slot='audienceScore')[0] #audience score
    critic_splat_audience_score_clean = [score.text.strip() for score in critic_splat_audience_score]
    critic_splat_audience_reviews = soup_critic_splat.find_all(slot='audienceReviews') #no. audience reviews
    critic_splat_audience_reviews_clean = [score.text.strip() for score in critic_splat_audience_reviews]
    return critic_splat_title_clean, critic_splat_info_clean, critic_splat_genre_clean, critic_splat_critic_score_clean, critic_splat_critic_reviews_clean, critic_splat_audience_score_clean, critic_splat_audience_reviews_clean
```

```python
critic_splat = critic_splat_general_info('https://rottentomatoes.com/m/a_working_man')
critic_splat
```

```python
critic_splat = [critic_splat_general_info(url) for url in full_links_splat]
critic_splat
```

```python
critic_splat_messy_df = pd.DataFrame({'Movie': critic_splat})
critic_splat_messy_df = pd.DataFrame(critic_splat_messy_df['Movie'].to_list(), columns = ['Movie', 'General Info', 'Genre', 'Critic Score', 'No. Critic Reviews', 'Audience Score', 'No. Audience Reviews'])
critic_splat_messy_df
```

```python
critic_splat_messy_df['Movie'] = critic_splat_messy_df['Movie'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_messy_df['General Info'] = critic_splat_messy_df['General Info'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_messy_df['Genre'] = critic_splat_messy_df['Genre'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_messy_df['Critic Score'] = critic_splat_messy_df['Critic Score'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_messy_df['No. Critic Reviews'] = critic_splat_messy_df['No. Critic Reviews'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_messy_df['Audience Score'] = critic_splat_messy_df['Audience Score'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_messy_df['No. Audience Reviews'] = critic_splat_messy_df['No. Audience Reviews'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_messy_df
```

```python
critic_splat_messy_df[['Rating', 'General Info']] = critic_splat_messy_df['General Info'].str.split(',', n=1, expand=True)
#critic_splat_df[['Movie Length', 'Release Date']] = critic_splat_df['General Info'].str.split(',', n=1, expand=True)
critic_splat_messy_df
```

```python
critic_splat_trial = critic_splat_messy_df
#critic_splat_trial['Movie Length'] = critic_splat_trial['General Info'].str.split().str[-2:]
critic_splat_trial[['Movie Length', 'Release Date']] = critic_splat_trial['General Info'].str.rsplit(',', n=1, expand=True)
#critic_splat_trial['General Info'] = critic_splat_trial['General Info'].str[0:-2]
#critic_splat_messy_df['Movie Length'] = critic_splat_messy_df['Movie Length'].astype(str).str.strip('[]').str.replace("'", "")
critic_splat_trial
```

```python
critic_splat_df = critic_splat_trial.drop('General Info', axis = 1)
critic_splat_df = critic_splat_df.rename(columns={'Movie Length': 'Release Date', 'Release Date': 'Move Length'})
critic_splat_df
```

```python
critic_splat_file = 'critic_splat.xlsx'
```

```python
critic_splat_df.to_excel(critic_splat_file)
```
